## 将来予測実験データの読み込み

過去実験のデータは一気に読み込むことができたが、将来予測実験はデータ量が大きすぎるために一気に読み込もうとするとメモリー不足となる。そこで、ある程度の年数ごとにdaskによるデータの読み込み手順の設定を行い、データを読み込む。

ファイルのパスを設定する。

```python
bdir = 'dias-data/FORP/FORP-NP10/version4/MIROC5_rcp26_ver4/nc_mon/'
```

2000年代のファイル名のリストを作成する。

```python
files = list(pathlib.Path(bdir).glob("200*/forp-np-v4_thetao_mn_*.nc"))
```

ファイルを読み込み、SSTを抽出し、データセットds1を削除する。念のためデータをファイルに保存する。

```python
ds1 = xr.open_mfdataset(files)
sst1 = ds1.thetao.isel(lev=0)

start = time.perf_counter()
sst1.to_netcdf("FORPv4_MIROC5_rcp26_thetao.2006-2009.nc")
end = time.perf_counter()
print('{:.2f}'.format((end-start)/60))
```

```python
del ds1, sst1
```
